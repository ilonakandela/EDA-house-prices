#### Dealing with missing values

```python
df.isna().sum()
```

```python
total_missing = df.isna().sum().sort_values(ascending=False)
percentages_missing = (df.isna().sum() / df.isna().count()).sort_values(ascending=False)
missing_df = pd.concat([total_missing, percentages_missing], axis=1, keys=["Total_Missing", "Percentages_Missing"])
missing_df.head(25)
```

#### Drop all columns with missing data

```python
df = df.drop(missing_df[missing_df["Total_Missing"] > 10].index, axis=1)
```

```python
df.shape
```

```python
df.isna().sum().max()
```

#### Full value before

```python
df["MasVnrArea"] = df["MasVnrArea"].fillna(method="bfill")
#df["MasVnrType"] = df["MasVnrType"].fillna(method="bfill")
df["Electrical"] = df["Electrical"].fillna(method="bfill")
```

```python
numeric_columns = df.select_dtypes(include=[np.number])

# Calculate correlation matrix
correlation_matrix = numeric_columns.corr()

# Select top 10 correlated features with "SalePrice"
cols = correlation_matrix.nlargest(10, "SalePrice")["SalePrice"].index

# Calculate correlation matrix for selected columns
cm = np.corrcoef(df[cols].values.T)

# Plot the heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(cm, annot=True, fmt=".2f", xticklabels=cols.values, yticklabels=cols.values)
plt.show()
```

#### Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_calc_df = df[["OverallQual", "GrLivArea", "GarageCars","TotalBsmtSF", "YearBuilt"]]
```

```python
vif_df = pd.DataFrame()

vif_df["Columns"] = vif_calc_df.columns
vif_df["VIF"] = [variance_inflation_factor(vif_calc_df.values, i) 
                for i in range(vif_calc_df.shape[1])]
```

```python
vif_df
```

```python
plt.figure(figsize=(12,5))
sns.barplot(y="VIF", x="Columns", data=vif_df)
plt.title("VIF")
plt.show()
```
